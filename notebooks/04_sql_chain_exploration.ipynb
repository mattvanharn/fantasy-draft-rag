Draft SQL-generation prompt string

````python
QUESTION = "Which players had ADP under 50 but finished outside top 100 in 2024?"
sample_data = get_sample_rows("player_seasons", n=3)

SQL_GENERATION_PROMPT_TEMPLATE = """You are an expert SQLite query writer for fantasy football analysis.

Database schema:
{schema}

Sample rows from player_seasons:
{sample_data}

Rules:
- Return ONLY SQL. No explanation. No markdown fences.
- Do NOT include reasoning text or <think> tags.
- Generate ONLY a single SELECT query.
- Use only tables/columns that exist in schema.
- Use player_display_name for player names.
- seasonal_fantasy_points is half-PPR points.
- adp is average draft position (lower means earlier draft pick).
- season is integer (e.g., 2024).
- position values are abbreviations (QB, RB, WR, TE, K, DST).
- Use ORDER BY for ranking questions (best/top/worst).
- Unless user asks otherwise, include LIMIT 20.
- If comparing players/seasons, include relevant identifying columns.

Question: {question}

SQL:"""

SQL_GENERATION_PROMPT_DRAFT = SQL_GENERATION_PROMPT_TEMPLATE.format(
    schema=schema,
    sample_data=sample_data,
    question=QUESTION,
)

print(SQL_GENERATION_PROMPT_DRAFT[:2500])
````

Call LLM to generate SQL from your draft prompt

````python
from langchain_groq import ChatGroq
import os
from dotenv import load_dotenv

load_dotenv()
llm_sql = ChatGroq(
    model=os.getenv("GROQ_MODEL", "llama-3.3-70b-versatile"),
    temperature=0.0,  # SQL generation should be deterministic
)

sql_out = llm_sql.invoke(SQL_GENERATION_PROMPT_DRAFT).content.strip()
print("Model:", os.getenv("GROQ_MODEL", "llama-3.3-70b-versatile"))
print(sql_out[:2000])
````

Clean + execute generated SQL

````python
# Robust extraction: handles <think> blocks, prose, and markdown fences.
import re


def extract_select_sql(raw: str) -> str:
    text = re.sub(r"(?is)<think>.*?</think>", "", raw).strip()
    text = text.replace("```sql", "").replace("```", "").strip()
    match = re.search(r"(?is)\bselect\b.*?(?:;|$)", text)
    if not match:
        raise ValueError("No SELECT statement found in model output")
    return match.group(0).strip().rstrip(";")


clean_sql = extract_select_sql(sql_out)
print(clean_sql)

rows = execute_query(clean_sql)
print(format_results(rows, max_rows=20))
````

## Part 2: How `sql_chain.py` fits with `database.py`

- **`database.py`** — Loads parquets into in-memory SQLite, exposes `get_schema`, `get_sample_rows`, `execute_query` (SELECT-only), `format_results`.
- **`sql_chain.py`** — Does **not** replace the database layer. It registers **three tools** that call into it:
  1. **`sql_db_list_tables`** — names of tables
  2. **`sql_db_schema`** — PRAGMA columns + sample rows (same DB as above)
  3. **`sql_db_query`** — runs `execute_query` + `format_results`

**`create_agent`** (from LangChain) lets the model decide *when* to list tables, read schema, and run SQL — similar in spirit to the [LangChain SQL agent tutorial](https://docs.langchain.com/oss/python/langchain/sql-agent).

Run the cells below in order: first call the tools yourself, then run the agent and read the **message trail**.

### Agent: one question, full message trail

`agent.invoke(...)` returns a dict with a `messages` list: Human → AI (optional `tool_calls`) → ToolMessage(s) → … → final AI answer.  
This is how you **debug** what the model actually did (which SQL it ran, etc.).

````python
import os

from dotenv import load_dotenv
from langchain.agents import create_agent
from langchain_groq import ChatGroq

from ff_ai_assistant.sql_chain import (
    SYSTEM_PROMPT,
    sql_db_list_tables,
    sql_db_schema,
    sql_db_query,
)

load_dotenv()
llm = ChatGroq(
    model=os.getenv("GROQ_MODEL", "llama-3.3-70b-versatile"),
    temperature=0.0,
)
agent = create_agent(
    model=llm,
    tools=[sql_db_list_tables, sql_db_schema, sql_db_query],
    system_prompt=SYSTEM_PROMPT,
)

question = "Who were the top 3 RBs by half-PPR points in 2024?"
result = agent.invoke({"messages": [{"role": "user", "content": question}]})

for i, msg in enumerate(result["messages"]):
    name = type(msg).__name__
    print(f"\n--- [{i}] {name} ---")
    content = getattr(msg, "content", None)
    if content:
        text = content if isinstance(content, str) else str(content)
        print(text[:1200] + ("..." if len(text) > 1200 else ""))
    tool_calls = getattr(msg, "tool_calls", None)
    if tool_calls:
        for tc in tool_calls:
            tname = tc.get("name") if isinstance(tc, dict) else getattr(tc, "name", None)
            args = tc.get("args") if isinstance(tc, dict) else getattr(tc, "args", None)
            print("  tool_call:", tname, args)
````

### Manual prompt path: extract SQL when the model adds `</think>` or prose

If you use a **single** `ChatGroq.invoke(...)` with a hand-written prompt, the reply may include reasoning tags or text before `SELECT`.  
`execute_query` only accepts strings that start with `SELECT` — so parse first:

```python
clean = extract_select_sql(sql_out)
rows = execute_query(clean)
```

````python
import re


def extract_select_sql(raw: str) -> str:
    """Strip common noise and return the first SELECT ... statement."""
    # Some chat models wrap hidden reasoning in <think>...</think> (or similar).
    text = re.sub(r"(?is)<think>.*?</think>", "", raw).strip()
    text = text.replace("```sql", "").replace("```", "").strip()
    m = re.search(r"(?is)\bselect\b.*?(?:;|$)", text)
    if not m:
        raise ValueError("No SELECT found in model output:\n" + raw[:500])
    return m.group(0).strip().rstrip(";")


# Example (uncomment after you have `sql_out` from an LLM cell):
# clean = extract_select_sql(sql_out)
# print(clean)
# rows = execute_query(clean)
# print(format_results(rows, max_rows=20))
````
